**PACKAGES**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**READ FILE**

```
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/movie_ratings.csv')
data.head()
```

```
data=data.drop(columns=['Unnamed: 0'])
data.head()
```

```
ratings= data.pivot_table(index=['user_id'],columns=['title'],values='rating')
ratings.head()
```

**CORRELATION**

```
correlation=ratings.corr()
correlation.head()
```

**MODIFY RESULT**


1.  No. of Users Rated < 100 :  Throw Out




```
correlation = ratings.corr(method='pearson', min_periods=100)
correlation.head()
```

**#######################RECOMMENDATION##########################**

**EXPLORE USER[0] RATINGS**

```
myRatings = ratings.loc[0].dropna()
myRatings
```

```
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    sims = correlation[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[i])
    simCandidates = simCandidates.append(sims)

simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

**Sum movie rating**, so that movies which came more than once gets more priority.

```
simCandidates = simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

**REMOVE ALREADY WATCHED**
Drop  movies which were in myRatings

```
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
